```python
import pandas as pd
```

```python
df = pd.read_csv("MeiLeaf_data.csv", sep = ",")
```

```python
df.index = df["Name"]
```

```python
relevant_data = df[df["Price per Gram"] < 50][df["Number of Reviews"] > 0][df["Score"]>0]
```

# Find the best teas by rating, giving more weight to the ones with many reviews

```python
base_rating = (relevant_data["Score"] * relevant_data["Number of Reviews"]).sum() / relevant_data["Number of Reviews"].sum()

import math
c = math.sqrt((relevant_data["Number of Reviews"]).mean())
```

```python
# do bayesian averaging
relevant_data["Bayesian Average"] = relevant_data.apply(lambda x: (x["Score"]*x["Number of Reviews"]+c*base_rating)/(x["Number of Reviews"]+c), axis = 1)
```

```python
# I'm willing to pay {amount} times as much for an increase in rating of {increase}
amount = 2
increase = 2

interesting_data = pd.DataFrame((relevant_data.index, relevant_data["Bayesian Average"], relevant_data["Score"], relevant_data["Number of Reviews"], relevant_data["Price per Gram"], relevant_data["Type"], relevant_data["Link"]), index = ["Name", "Bayesian Average", "Original Rating", "Number of Reviews", "Price per Gram", "Type", "Link"]).transpose()
interesting_data["Final Score"] = ((interesting_data["Bayesian Average"]).apply(lambda x: math.pow(amount, 1/increase)**x)/interesting_data["Price per Gram"]).apply(math.log)
```

```python
final_data = interesting_data.sort_values(by = "Final Score", ascending = False).reset_index(drop = True)
final_data.to_csv("final_data.csv", sep = ";", index = False)
```
